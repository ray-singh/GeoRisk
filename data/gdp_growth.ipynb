```python
import sdmx 
import pandas as pd

# Establishing connection to IMF
IMF_DATA = sdmx.Client('IMF_DATA')

# Code for accessing GDP growth dataset
key = '.NGDP_RPCH.A'

# Importing dataset as a dataframe
dataset = pd.read_csv('dataset.csv')
```

```python
gdp_growth_rate = []

for id in dataset['id']:
    temp_key = id + key
    try:
        data_msg = IMF_DATA.data('WEO',  key=temp_key, params={'startPeriod': 2023, 'endPeriod':2025})
        gdp_df = sdmx.to_pandas(data_msg)
        gdp_growth_rate.append(gdp_df[-1])
    except:
        gdp_growth_rate.append(-99999.0)
```

```python
dataset['gdp_percent_change'] = gdp_growth_rate
dataset.head()
```

```python
dataset.to_csv('dataset.csv', index=False)
```
