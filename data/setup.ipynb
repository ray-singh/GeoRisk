```python
import wbdata
import pandas as pd
```

```python
df = pd.DataFrame(columns=['id', 'iso2Code', 'name'])
```

```python
for country in wbdata.get_countries():
    new_row = {'id':country['id'], 'iso2Code': country['iso2Code'], 'name':country['name']}
    df.loc[len(df)] = new_row
```
