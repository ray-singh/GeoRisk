```python
import sdmx 
import pandas as pd

# Establishing connection to IMF
IMF_DATA = sdmx.Client('IMF_DATA')

# Code for accessing current account balance as a % of GDP
key = '.BCA_NGDPD.A'

# Importing dataset as a dataframe
dataset = pd.read_csv('dataset.csv')
```

```python
cab = []

for id in dataset['id']:
    temp_key = id + key
    try:
        data_msg = IMF_DATA.data('WEO',  key=temp_key, params={'startPeriod': 2023, 'endPeriod':2025})
        cab_df = sdmx.to_pandas(data_msg)
        cab.append(cab_df[-1])
    except:
        cab.append(-99999.0)
```

```python
dataset['current_account_balance_%gdp'] = cab
dataset
```

```python
dataset.to_csv('dataset.csv', index=False)
```
