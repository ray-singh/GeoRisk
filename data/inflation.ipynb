# Procesing Inflation Data From IMF'S CPI Dataset

The goal of this notebook is to fetch and process inflation data from IMF'S API, adding each country's latest period average percent change into the dataset. 

Syntax of key used for API call: <Country_code>.CPI.<Category_id>.<Transformation_type>.<Frequency>

```python
import sdmx 
import pandas as pd

# Establishing connection to IMF
IMF_DATA = sdmx.Client('IMF_DATA')

# Code for accessing CPI dataset
key = 'CPI._T.YOY_PCH_PA_PT.Q'

# Importing dataset as a dataframe
dataset = pd.read_csv('dataset.csv')
```

```python
inflation = []

for id in dataset['id']:
    temp_key = id + '.' + key
    try:
        data_msg = IMF_DATA.data('CPI',  key=temp_key, params={'startPeriod': 2023, 'endPeriod':2025})
        cpi_df = sdmx.to_pandas(data_msg)
        inflation.append(cpi_df[-1])
    except:
        inflation.append(-99999.0)
```

```python
dataset['inflation_rate'] = inflation
```
